```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

Вычислим скользящее среднее температуры с окном в 30 дней для сглаживания краткосрочных колебаний.

```python
# Загружаем данные
df = pd.read_csv('temperature_data.csv', parse_dates=['timestamp'])

# Вычислим скользящее среднее температуры с окном в 30 дней с группировкой по городам
df['moving_avg'] = df.groupby('city')['temperature'].transform(lambda x: x.rolling(window=30, min_periods=1).mean())
```

Рассчитаем среднюю температуру и стандартное отклонение для каждого сезона в каждом городе.

```python
# Рассчитаем среднюю температуру и стандартное отклонение для каждого сезона в каждом городе
seasonal_stats = df.groupby(['city', 'season'])['temperature'].agg(['mean', 'std']).reset_index()

# Объединим сезонные статистики с основным DataFrame для дальнейшего анализа
df = df.merge(seasonal_stats, on=['city', 'season'], suffixes=('', '_seasonal'))
```

```python
# Выявим аномалии, где температура выходит за пределы среднего ±2σ
df['is_anomaly'] = (df['temperature'] < (df['mean'] - 2 * df['std'])) | (df['temperature'] > (df['mean'] + 2 * df['std']))
df.head(3)
```

```python
%%time
df['moving_avg'] = df.groupby('city')['temperature'].transform(lambda x: \
                                    x.rolling(window=30, min_periods=1).mean())
seasonal_stats = df.groupby(['city', 'season'])['temperature']. \
                                             agg(['mean', 'std']).reset_index()
df = df.merge(seasonal_stats, on=['city', 'season'], suffixes=('', '_seasonal'))
df = df.drop(['mean_seasonal', 'std_seasonal'], axis=1)
df['is_anomaly'] = (df['temperature'] < \
(df['mean'] - 2 * df['std'])) | (df['temperature'] > (df['mean'] + 2 * df['std']))
```

Теперь замерим время выполнения того же кода с распараллеливанием с помощью polars

```python
%%time
# Преобразуем pandas в polars DataFrame
pl_df = pl.from_pandas(df)

# Вычисление скользящего среднего
pl_df = pl_df.with_columns(
    pl.col("temperature")
    .rolling_mean(window_size=2, min_periods=1)
    .alias("moving_avg")
)

# Вычисление статистики по сезонам
seasonal_stats = (
    pl_df.group_by(["city", "season"])
    .agg([
        pl.col("temperature").mean().alias("mean"),
        pl.col("temperature").std().alias("std")
    ])
)

# Объединение данных с сезонной статистикой
pl_df = pl_df.join(seasonal_stats, on=["city", "season"])

# Вычисление и отметка аномалий
pl_df = pl_df.with_columns(
    (
        (pl.col("temperature") < (pl.col("mean") - 2 * pl.col("std"))) |
        (pl.col("temperature") > (pl.col("mean") + 2 * pl.col("std")))
    ).alias("is_anomaly")
)
```

Получилось быстрее примерно в 2 раза

2. **Мониторинг текущей температуры**:

Получим текущую температуру для Москвы через OpenWeatherMap API.

```python
import requests

def get_temperature(city_name, api_key):
    base_url = "https://api.openweathermap.org/data/2.5/weather"

    params = {
        'q': city_name,
        'appid': api_key,
        'units': 'metric'
    }

    response = requests.get(base_url, params=params)
    data = response.json()

    if response.status_code == 200:
        temperature = data['main']['temp']
        return temperature
    else:
        print(f"Ошибка: {data.get('message', 'Неизвестная ошибка')}")
        return None

api_key = ""
city_name = "Moscow"
temperature = get_temperature(city_name, api_key)

print(f"Температура в городе {city_name}: {temperature}°C")
```

Определим, является ли текущая температура нормальной, исходя из исторических данных для текущего сезона. Буду сравнивать с медианное значение исторической температуры на эту дату, если она не аномальная

```python
import pandas as pd

def compare_temperature(date, actual_temp, city, df):
    date = pd.to_datetime(date)

    # Фильтрация данных для конкретного города и указанной даты и без аномалий
    city_data = df[(df['city'] == city) &
                   (df['timestamp'].dt.month == date.month) &
                   (df['timestamp'].dt.day == date.day) &
                   (df['is_anomaly'] == False)]

    # print(city_data)

    if city_data.empty:
        return "Нет данных для указанного города и даты."

    # Возьмём медианное значение исторической температуры на эту дату,
    # если она не аномальная
    mean_temp = city_data['temperature'].median()
    # mean_temp = city_data['mean'].iloc[0] # менее точный вариант с полученным ранее сезонным средним
    std_temp = city_data['std'].iloc[0]

    # Вычисление отклонения фактической температуры от средней
    deviation = actual_temp - mean_temp
    # Определение, насколько фактическая температура отличается от исторической
    z_score = deviation / std_temp if std_temp != 0 else float('inf')

    # Вывод на основе Z-оценки
    if abs(z_score) < 1:
        assessment = "Температура близка к историческому среднему, не является аномалией."
    elif abs(z_score) < 2:
        assessment = '''Температура находится в пределах одного стандартного
    отклонения от среднего, что является нормальным размахом для таких данных.'''
    else:
        assessment = '''Температура считается аномальной относительно исторических
         данных — слишком холодной или слишком теплой.'''

    result = (
        f"Результат сравнения текущей температуры с историческими данными для города {city}\n"
        f"Фактическая температура: {actual_temp}°C\n"
        f"Средняя историческая температура для {city} на эту дату: {round(mean_temp, 2)}°C\n"
        f"Стандартное отклонение: {round(std_temp, 2)}°C\n"
        f"Отклонение фактической температуры: {round(deviation, 2)}°C\n"
        f"Z-оценка: {z_score:.2f}\n"
        f"Вывод: {assessment}\n"
        f"{'*' * 80}\n"
    )

    return result

sample_date = "2024-12-19"
sample_temp = temperature
sample_city = city_name

print(compare_temperature(sample_date, sample_temp, sample_city, df))
```

```python
%%time
import aiohttp
import asyncio

async def get_temperature_async(city_name, api_key):
    base_url = "https://api.openweathermap.org/data/2.5/weather"

    params = {
        'q': city_name,
        'appid': api_key,
        'units': 'metric'
    }

    async with aiohttp.ClientSession() as session:
        async with session.get(base_url, params=params) as response:
            data = await response.json()

            if response.status == 200:
                temperature = data['main']['temp']
                return temperature
            else:
                print(f"Ошибка: {data.get('message', 'Неизвестная ошибка')}")
                return None

async def main():
    api_key = ""
    city_name = "Moscow"

    for city_name in df['city'].unique():
        # Получение температуры для города
        temperature = await get_temperature_async(city_name, api_key)

        # Вызов функции сравнения температуры
        print(compare_temperature(sample_date, temperature, city_name, df))

# Запуск события
asyncio.run(main())
```
